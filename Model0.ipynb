```python
col_names = ['budget','revenue','popularity','weighted_rating','director_Rank','collection_Rank','prod_Comp_Rank']
fig, ax = plt.subplots(len(col_names), figsize=(12,36))

for i, col_val in enumerate(col_names):

    sns.distplot(df[col_val], hist=True, ax=ax[i])
    ax[i].set_title('Freq dist '+col_val, fontsize=10)
    ax[i].set_xlabel(col_val, fontsize=8)
    ax[i].set_ylabel('Count', fontsize=8)

plt.show()
```

```python
df.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8); 
plt.show()
```

```python
#correlation matrix is a table showing the value of the correlation coefficient 
#(Correlation coefficients are used in statistics to measure how strong a relationship is between two variables. ) between sets of variables. 
#Each attribute of the dataset is compared with the other attributes to find out the correlation coefficient. 
#This analysis allows you to see which pairs have the highest correlation, the pairs which are highly correlated represent the same variance of the dataset thus we can further analyze them to understand which attribute among the pairs are most significant for building the model.

f, ax = plt.subplots(figsize=(15, 12))
corr = df.corr()
sns.heatmap(corr,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
plt.show()
```

```python

import statsmodels.api as sm

X = df["budget"]
y = df["revenue"]

# Note the difference in argument order
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model

# Print out the statistics
model.summary()
```
